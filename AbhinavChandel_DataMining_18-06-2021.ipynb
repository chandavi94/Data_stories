### Training and Testing of data

```python
x = ins_df.drop("Claimed", axis=1)

y = ins_df.pop("Claimed")

x.head()
```

```python
y.head()
```

```python
from scipy.stats import zscore
x_scaled=x.apply(zscore)
x_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.30, random_state= 0)
```

```python
print('x_train',x_train.shape)
print('x_test',x_test.shape)
print('y_train',y_train.shape)
print('y_test',y_test.shape)
```

## Decision Tree Formation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_dtcl = {
    'criterion': ['gini'],
    'max_depth': [4,10,12,30,50],
    'min_samples_leaf': [50,100,150], 
    'min_samples_split': [150,300,450],
}

dtcl = DecisionTreeClassifier(random_state=0)

grid_search_dtcl = GridSearchCV(estimator = dtcl, param_grid = param_grid_dtcl, cv = 10)
```

```python
grid_search_dtcl.fit(x_train, y_train)
print(grid_search_dtcl.best_params_)
best_grid_dtcl = grid_search_dtcl.best_estimator_
best_grid_dtcl
```

```python
from sklearn import tree
train_char_label = ['no', 'yes']
tree_reg = open('tree_reg.dot','w')
dot_data = tree.export_graphviz(best_grid_dtcl, out_file= tree_reg ,
                                feature_names = list(x_train),
                                class_names = list(train_char_label))

tree_reg.close()
dot_data
```

```python
print (pd.DataFrame(best_grid_dtcl.feature_importances_, columns = ["Importance"], 
                    index = x_train.columns).sort_values('Importance',ascending=False))
```

```python
ytrain_predict = best_grid_dtcl.predict(x_train)
ytest_predict = best_grid_dtcl.predict(x_test)
```

```python
ytest_predict_prob = best_grid_dtcl.predict_proba(x_test)
pd.DataFrame(ytest_predict_prob).head()
```

### Random Forest Creation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score,roc_curve,classification_report,confusion_matrix
```

```python
param_grid_rfcl = {
    'max_depth': [4,5,10],
    'max_features': [2,5,6],
    'min_samples_leaf': [8,10,12],
    'min_samples_split': [45,50,55],
    'n_estimators': [100,200,300]
}

rfcl = RandomForestClassifier(random_state=0)

grid_search_rfcl = GridSearchCV(estimator = rfcl, param_grid = param_grid_rfcl, cv = 5)
```

```python
grid_search_rfcl.fit(x_train, y_train)
print(grid_search_rfcl.best_params_)
best_grid_rfcl = grid_search_rfcl.best_estimator_
best_grid_rfcl
```

```python
ytrain_predict_rfcl = best_grid_rfcl.predict(x_train)
ytest_predict_rfcl = best_grid_rfcl.predict(x_test)
```

```python
ytest_predict_prob_rfcl=best_grid_rfcl.predict_proba(x_test)
pd.DataFrame(ytest_predict_prob_rfcl).head()
```

```python
print (pd.DataFrame(best_grid_rfcl.feature_importances_, columns = ["Importance"], 
                    index = x_train.columns).sort_values('Importance',ascending=False))
```

### Neural Network Creartion

```python
from sklearn.neural_network import MLPClassifier

param_grid_clf = {
    'hidden_layer_sizes': [50,200],
    'max_iter': [2500,4000,5000],
    'solver': ['adam','sgd'],
    'tol': [0.01,0.001], 
}

clf = MLPClassifier(random_state=0)

grid_search_clf = GridSearchCV(estimator = clf, param_grid = param_grid_clf, cv = 10)
```

```python
grid_search_clf.fit(x_train, y_train)
grid_search_clf.best_params_
best_grid_clf = grid_search_clf.best_estimator_
best_grid_clf
```
